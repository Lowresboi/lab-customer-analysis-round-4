```python
# Importing the libraries:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sync the drive:
from google.colab import drive # if you are using google colab and drive
drive.mount('/content/drive')
```

```python
# Adding the content for the Lab from my drive:

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Labs 3/lab-customer-analysis-round-3/files_for_lab/csv_files/marketing_customer_analysis.csv')
data
```

```python
# Showing the data info:

data.info()
```

```python
# Describing the data:

data.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Response', data=data)
plt.title('Total Number of Responses')
plt.show()
```

```python
# Plot response rate by sales channel:
plt.figure(figsize=(10, 6))
sns.countplot(x='Sales Channel', hue='Response', data=data)
plt.title('Response Rate by Sales Channel')
plt.show()
```

```python
# Plot response rate by total claim amount:
plt.figure(figsize=(12, 6))
sns.boxplot(x='Response', y='Total Claim Amount', data=data)
plt.title('Response Rate by Total Claim Amount')
plt.show()
```

```python
# Plot response rate by income:
plt.figure(figsize=(12, 6))
sns.boxplot(x='Response', y='Income', data=data)
plt.title('Response Rate by Income')
plt.show()
```

```python
# Checking the types of the columns:

numerical = data.select_dtypes("number")
categorical = data.select_dtypes("object")
```

```python
# Showing the results:

numerical
```

```python
# Visualizing the numericals:

for column in numerical.columns:
    sns.histplot(data=numerical, x=column, kde=True)
    plt.show()
```

```python
# Checking the multicollinearity:

correlation_matrix = numerical.corr()
correlation_matrix
```

```python
# Drop one of the two features that show a high correlation between them (greater than 0.9):
high_correlation_features = np.where(correlation_matrix > 0.9)
high_correlation_features
```

```python
# Showing it with a heat map:

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```
